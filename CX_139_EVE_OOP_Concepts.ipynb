```python
# constructor __init__() : is class method used to initialise object (setting starting value of attributes)
# constructor is invoked automatically when object is created
```

```python
class employees:
  def __init__(self,id,name,sal):
    self.empid=id
    self.empname=name
    self.salary=sal
  def display(self):
    print("Employee Id=",id)
    print("Employee Name=",self.empname)
    print("Employee Salary=",self.salary)

```

```python
# create object of class employees
eobj=employees(101,'Namo',78520)# will invoke constructor (__init__())
eobj.display()
```

```python
# create class BankAccount  accholder, balance, define constructor, display

class BankAccounts:
  def __init__(self,accholder, balance):
    self.accholder=accholder
    self.balance=balance
  def display(self):
    print("accholder=",self.accholder)
    print("balance=",self.balance)
  def deposit(self,amt):
    self.balance=self.balance+amt
    print("Amount deposited Successfully, Avail.Bal is ",self.balance)
  def withdraw(self,amt):
    if amt<self.balance:
      self.balance=self.balance-amt
      print("Amount withdrawn ",amt," Avail.Bal is ",self.balance)
    else:
      print("Insufficient Funds")

b1=BankAccounts('Satyendra',1000)
b1.display()
```

```python
b2=BankAccounts("Ajay",1500)
b3=BankAccounts("Sushma",2000)
```

```python
objlist=[b1,b2,b3]
```

```python
for i in objlist:
  i.display()
```

Inheritance

```python
class vehicle:
  def __init__(self,m,f):
    self.manufacturer=m
    self.fueltype=f
  def display(self):
    print("Manufacturer=",self.manufacturer)
    print("Fuel type=",self.fueltype)
```

```python
# create a class called two-wheeler having attributes like maufacturer, fueltype and braketype

class two_wheeler(vehicle):
  def __init__(self,ma,fuel,braketype):
    super().__init__(ma,fuel)# initialising parent class constructor
    self.braketype=braketype
  def showbikes(self):
    super().display()
    print("Brake type=",self.braketype)
```

```python
av=two_wheeler("Honda","Petrol","Disk")
av.showbikes()
```
